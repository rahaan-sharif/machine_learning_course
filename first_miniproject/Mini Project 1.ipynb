```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load and Prepare Data

```python
data=pd.read_csv('googleplaystore.csv')
print("length: ", len(data))
```

## EDA

```python
columns=data.columns
print("columns length:  ", len(columns))
print("salam")
```

```python
#length of the table:
length=len(data[columns[0]])
print("length of the table:", length)
```

```python
nan_list=[]
for i in range(0, 13):
    for j in range(0, 10841):
        if(pd.isna(data[columns[i]][j])):
            nan_list+=[j]
print("all necessary values added to nan_list")
print("nan_list length:  ", len(nan_list))
```

```python
nan_list=list(set(nan_list))
print("all repetitive values deleted from nan_list")
print("nan_list length: ", len(nan_list))
```

```python
for i in range(0, 13):
    for j in range(0, len(nan_list)):
        data[columns[i]].pop(j)
print("all null values deleted from the table")
print("data lenght:  ", len(data[columns[0]]))
```
